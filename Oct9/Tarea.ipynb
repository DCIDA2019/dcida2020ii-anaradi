```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy
from scipy.stats import binom
from scipy.stats import poisson
```

```python
mu=70
var=2
sigma=math.sqrt(2)
N=100000
```

```python
data2=np.random.rand(N)
x2=np.sort(data2)
```

```python
data = np.random.normal(mu, sigma, N)
x = np.sort(data)
```

```python
#cdf con la distribución aleatoria normal y la distribución normal
y = (np.sqrt(2 *var * np.pi)**(-1))*np.exp(-0.5 * (1 / var) * (x - mu)**2)
y = y.cumsum()/(2*N)
F2 = np.array(range(N))/float(N)
```

```python
plt.plot(x, y) 
plt.show()
```

```python
#CDF con la distribución normal aleatoria 
plt.plot(x, F2) 
plt.show() 
```

```python
hx, hy, _ = plt.hist(x, bins=50)

plt.ylim(0.0,max(hx)+100)
plt.grid()
```
